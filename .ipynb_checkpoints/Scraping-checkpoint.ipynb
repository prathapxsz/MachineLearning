```python
import requests
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords

#Getting data from a source, In this case from wikipedia
url = 'https://en.wikipedia.org/wiki/Machine_learning'
r = requests.get(url)

#scraping data from html to text data using beautifulSoup library
soup = BeautifulSoup(r.text,'html5lib')
ptagData = soup.findAll('p')
str_data = ''
for data in ptagData:
    str_data += data.text

#Cleaning data using regex library
clean_data = re.sub(r'\[[0-9]*\]', ' ',str_data)
clean_data = re.sub(r'[^a-zA-Z]', ' ', clean_data)
clean_data = re.sub(r'\s+', ' ', clean_data)
str = clean_data.split()

#Removing stopwords from obtained data list
imp_words = []
for word in str:
    if word not in stopwords.words("english"):
        imp_words.append(word)
#print(imp_words)

#Getting highly repeated 20 words using nltk.FreqDist
freq_data = nltk.FreqDist(imp_words)
freq_data.plot(20)
```
