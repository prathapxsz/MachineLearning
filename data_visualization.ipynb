```python
import matplotlib.pyplot as plt
```

```python
x = [2,9,3,7]      # To represent distance
y = [7, 15, 5, 9]   # To represent Time
plt.xlabel("Distance")    # To represent label for x
plt.ylabel("Time")      # To represent label for y
plt.plot(x,y)         # To plot graph
plt.scatter(x,y,s=100, label="Railway Station") # To increase sizes of points
plt.legend()                 # To represent labels
plt.grid()             #To insert a grid 
```

```python
players = ['Dhoni', 'Virat', 'Rohit', 'Hardik']
scores = [32, 22, 12, 122]
plt.bar(players, scores, label="2020 scores of indian team")# To plot bar graphs
plt.legend()
```

```python
plt.pie(scores, labels=players, explode=(0,0,0,0.2), shadow= True, autopct= '%1.1ff%%')
plt.legend()
plt.show()
```
